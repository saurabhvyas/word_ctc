```python
# read all the hindi text 

with open('/home/saurabh/Documents/pytorch_gan/test/hindi.txt', 'r') as myfile:
    data=myfile.read() #.replace('\n', '')
```

```python
print(data)
```

```python
#text = text.decode("utf-8")

#this breaks the text on the white spaces, yielding a list of words:
words = data.split()
print(len(words))

# remove duplicate items from the list

words=list(set(words))
print(len(words))
```

```python
# write each item/word in words list to words_hindi.txt

with open ('/home/saurabh/Documents/tf_orange/tf_orange/words/hindi_words.txt', 'w') as fo:
   for word in words:
     fo.write(str(word) + '\n')
```
